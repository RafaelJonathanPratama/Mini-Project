```python
import matplotlib.pyplot as plt

# Data prevalensi CRD global (dalam juta kasus)
tahun = ['2010','2025']
prevalensi = [389.7, 544.9]

# Membuat pie chart
plt.figure(figsize=(7, 7))
plt.pie(prevalensi, labels=tahun, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff6666'])
plt.title('Tren Prevalensi Penyakit Pernapasan Kronis Global (2010 vs 2025)')
plt.axis('equal')  # Membuat pie chart berbentuk lingkaran sempurna
plt.show()
```
