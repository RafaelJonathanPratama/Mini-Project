```python
import matplotlib.pyplot as plt
import pandas as pd

# Data dummy
data = {
    'Komponen': ['CPU', 'RAM', 'Harddisk', 'GPU', 'Motherboard'],
    'Tahun_1990': [1, 4, 40, 0, 1],
    'Tahun_Sekarang': [32, 32, 2000, 12, 1]
}

df = pd.DataFrame(data)

x = range(len(df['Komponen']))

plt.figure(figsize=(10, 6))
plt.bar([p - 0.2 for p in x], df['Tahun_1990'], width=0.4, label='1990')
plt.bar([p + 0.2 for p in x], df['Tahun_Sekarang'], width=0.4, label='Sekarang')

plt.xticks(x, df['Komponen'])
plt.ylabel('Kapasitas (misalnya: GB atau Core)')
plt.title('Perbandingan Komponen Komputer: 1990 vs Sekarang')
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
