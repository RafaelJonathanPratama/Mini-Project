```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data dummy
data = {
    'Country': ['USA', 'India', 'China', 'Brazil', 'Nigeria'] * 2,
    'Year': [1990]*5 + [2025]*5,
    'DeathRate': [15, 12, 10, 8, 7, 18, 20, 16, 14, 19],
    'Population': [250, 870, 1140, 150, 95, 330, 1400, 1420, 210, 230]
}

# Membuat dataframe
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Population',
    y='DeathRate',
    hue='Year',
    style='Year',
    s=100,
    palette='Set1'
)

# Memberi label
for i in range(len(df)):
    plt.text(df['Population'][i]+10, df['DeathRate'][i], df['Country'][i], fontsize=9)

plt.title('Angka Kematian Akibat Diabetes (1990 vs 2025)', fontsize=14)
plt.xlabel('Populasi (juta)')
plt.ylabel('Angka Kematian per 100.000 penduduk')
plt.grid(True)
plt.tight_layout()
plt.legend(title='Tahun')
plt.show()
```
