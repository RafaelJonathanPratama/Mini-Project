```python
import matplotlib.pyplot as plt
import pandas as pd

# Data dummy
data = {
    'Tahun': [1890, 1890, 1890, 1990, 1990, 1990],
    'Penyakit': ['TBC', 'Kolera', 'Malaria', 'TBC', 'Kolera', 'Malaria'],
    'JumlahKasus': [800000, 600000, 900000, 300000, 100000, 500000]
}

df = pd.DataFrame(data)

# Pivot data untuk line chart
pivot_df = df.pivot(index='Penyakit', columns='Tahun', values='JumlahKasus').reset_index()

# Plot
plt.figure(figsize=(10, 6))
for index, row in pivot_df.iterrows():
    plt.plot([1890, 1990], [row[1890], row[1990]], marker='o', label=row['Penyakit'])

plt.title('Jumlah Kasus Penyakit Menular (1890 vs 1990)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kasus')
plt.legend(title='Penyakit')
plt.grid(True)
plt.tight_layout()
plt.show()
```
